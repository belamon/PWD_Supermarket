#### what is RFM?

#### Recency: is a measure how many days ago a customer made their last purchase. It matters because a customer who bought something last week will respond better to ads more than the one that shopped 6 months ago. 

#### Frequency represents how often a customer makes purchases. We can calculate it as the sum of purchases across all channel

#### Monetary represnet the total spending by a customer. High spending customers is more valuable than low spending customers.

```python
# Calculate Total Frequency and Total Monetary Value
df['Frequency'] = df[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum(axis=1)
df['Monetary'] = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)

# RFM columns (Recency, Frequency, Monetary)
rfm_data = df[['ID', 'Recency', 'Frequency', 'Monetary']]
```

#### now lets score customers on each metic 
#### 1: lowest value means least recent, least frequency
#### 5: highest value (most recent, most frequenct, highest moneayr)

```python
# Assign scores to Recency (lower is better, so reverse ranking)
rfm_data['Recency_Score'] = pd.qcut(rfm_data['Recency'], q=5, labels=[5, 4, 3, 2, 1])

# Assign scores to Frequency and Monetary (higher is better)
rfm_data['Frequency_Score'] = pd.qcut(rfm_data['Frequency'], q=5, labels=[1, 2, 3, 4, 5])
rfm_data['Monetary_Score'] = pd.qcut(rfm_data['Monetary'], q=5, labels=[1, 2, 3, 4, 5])
```

### now we concatenate the recency score, frequency score, and monetary score tables to create a composite RFM Score

```python
# Create RFM Score
rfm_data['RFM_Score'] = rfm_data['Recency_Score'].astype(str) + rfm_data['Frequency_Score'].astype(str) + rfm_data['Monetary_Score'].astype(str)

# Display the first few rows
print(rfm_data.head())
```

#### Now lets make customer segment based on RFM scores
#### we make segment based on RFM such as 
#### champions: Recent, frequent and high spendders (555): we can offer exclusive pers here and do marketing strategy that make it more pesonal. 
#### loyal customers: frequent and high spenders but not the most recent (455,545 ) the type of marketing works for this is to encourage discount through referrals
#### at risk: previously high value customer with high recency (155,255). Send win-back campaigns with personalized offers or discounts to re-engage
#### others: for this category we can use general campaign to re- engage them



```python
# Define segments
def segment_customer(row):
    if row['RFM_Score'] == '555':
        return 'Champion'
    elif row['Recency_Score'] >= 4 and row['Frequency_Score'] >= 3:
        return 'Loyal Customer'
    elif row['Recency_Score'] <= 2 and row['Frequency_Score'] >= 3:
        return 'At-Risk'
    elif row['Monetary_Score'] >= 4:
        return 'Big Spender'
    else:
        return 'Other'

# Apply segmentation
rfm_data['Segment'] = rfm_data.apply(segment_customer, axis=1)

# View segment distribution
print(rfm_data['Segment'].value_counts())
```
